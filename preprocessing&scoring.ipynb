# 데이터 전처리

```python
result = result73
```

```python
result['코멘트']
```

```python
# 불필요한 문자 제거
pattern1 = '[-=+,#/\?:;^$.@*"~%!\\\n\r\t▼♣§☆♡´∀｀ﾉ♥▲�▶＊━“”]'

text = re.sub(pattern1, '', result['코멘트'][0]).strip()
print(text)
```

```python
# 데이터프레임 적용
result['코멘트'] = result['코멘트'].progress_map(lambda x:re.sub(pattern1, '', x).strip())
```

```python
# 1차 전처리
result['코멘트'].head(15)
```

```python
id_list = result['아이디']
```

```python
# 댓글의 인스타그램 아이디 1차 제거
for i in range(len(id_list)):
  target = id_list[i]
  for j in range(len(result['코멘트'])):
    new = re.sub(target, "", result["코멘트"][j])
    result["코멘트"][j] = new
```

```python
# 2차 전처리
result["코멘트"].head(15)
```

```python
# 댓글 내 언급된 A / B / AB / X 분류
only_a = result[(result['코멘트'].str.lower().str.contains('a')) & ~(result['코멘트'].str.lower().str.contains('b'))]
only_b = result[~(result['코멘트'].str.lower().str.contains('a')) & (result['코멘트'].str.lower().str.contains('b'))]
both_ab = result[(result['코멘트'].str.lower().str.contains('a')) & (result['코멘트'].str.lower().str.contains('b'))]
none = result[~(result['코멘트'].str.lower().str.contains('a')) & ~(result['코멘트'].str.lower().str.contains('b'))]
```

```python
# A 만 언급한 댓글 모음
only_a
```

```python
# 불필요한 댓글 제거
num = [5]
target_dropped = []

for i in range(len(num)):
  target_dropped.append(only_a.index[num[i]])

only_a_dropped = only_a.drop(target_dropped)

if len(num) == 0:
  target_dropped = []
  only_a_dropped = only_a.drop(target_dropped)

# 3차 전처리
only_a_dropped
```

```python
# B 만 언급한 댓글 모음
only_b
```

```python
# 불필요한 댓글 제거
num = []
target_dropped = []

for i in range(len(num)):
  target_dropped.append(only_b.index[num[i]])

only_b_dropped = only_b.drop(target_dropped)

if len(num) == 0:
  target_dropped = []
  only_b_dropped = only_b.drop(target_dropped)

# 3차 전처리
only_b_dropped
```

# 감성 점수 계산

```python
from pororo import Pororo
```

```python
sa = Pororo(task="sentiment", model="brainbert.base.ko.shopping", lang="ko")
# sa = Pororo(task="sentiment", model="brainbert.base.ko.nsmc", lang="ko")
```

```python
# A 댓글 별 긍정/부정 감성 점수
only_a_dropped['A score'] = [None] * len(only_a_dropped)
only_a_dropped['B score'] = [None] * len(only_a_dropped)

for i in range(len(only_a_dropped)):
  print(only_a_dropped['코멘트'].iloc[i])
  score = sa(only_a_dropped['코멘트'].iloc[i], show_probs=True)
  print(score)
  only_a_dropped['A score'][only_a_dropped.index[i]] = max([score['positive'], score['negative']]) * 2
  only_a_dropped['B score'][only_a_dropped.index[i]] = min([score['positive'], score['negative']]) * 2

only_a_dropped
```

```python
wanted = [0,3,6]

for i in range(len(wanted)):
  only_a_dropped['A score'][only_a_dropped['A score'].index[wanted[i]]] = 1
  only_a_dropped['B score'][only_a_dropped['B score'].index[wanted[i]]] = 0

only_a_dropped
```

```python
# B 댓글 별 긍정/부정 감성 점수
only_b_dropped['A score'] = [None] * len(only_b_dropped)
only_b_dropped['B score'] = [None] * len(only_b_dropped)

for i in range(len(only_b_dropped)):
  print(only_b_dropped['코멘트'].iloc[i])
  score = sa(only_b_dropped['코멘트'].iloc[i], show_probs=True)
  print(score)
  
  only_b_dropped['A score'][only_b_dropped.index[i]] = min([score['positive'], score['negative']]) * 2
  only_b_dropped['B score'][only_b_dropped.index[i]] = max([score['positive'], score['negative']]) * 2

only_b_dropped
```

```python
wanted = [1,2,5]

for i in range(len(wanted)):
  only_b_dropped['A score'][only_b_dropped['A score'].index[wanted[i]]] = 0
  only_b_dropped['B score'][only_b_dropped['B score'].index[wanted[i]]] = 1

only_b_dropped
```

```python
# AB 댓글 별 긍정/부정 감성 점수
both_ab_dropped['Primary score'] = [None] * len(both_ab_dropped)
both_ab_dropped['Secondary score'] = [None] * len(both_ab_dropped)

for i in range(len(both_ab_dropped)):
  print(both_ab_dropped['코멘트'].iloc[i])
  score = sa(both_ab_dropped['코멘트'].iloc[i], show_probs=True)
  print(score)
  both_ab_dropped['Primary score'][both_ab_dropped.index[i]] = max([score['positive'], score['negative']]) * 2
  both_ab_dropped['Secondary score'][both_ab_dropped.index[i]] = min([score['positive'], score['negative']]) * 2

both_ab_dropped
```

```python
# X 댓글 별 긍정/부정 감성 점수
none_dropped['Primary score'] = [None] * len(none_dropped)
none_dropped['Secondary score'] = [None] * len(none_dropped)

for i in range(len(none_dropped)):
  print(none_dropped['코멘트'].iloc[i])
  score = sa(none_dropped['코멘트'].iloc[i], show_probs=True)
  print(score)
  none_dropped['Primary score'][none_dropped.index[i]] = max([score['positive'], score['negative']]) * 2
  none_dropped['Secondary score'][none_dropped.index[i]] = min([score['positive'], score['negative']]) * 2

none_dropped
```
